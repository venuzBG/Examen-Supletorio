# Splines cúbicos

```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos")
plt.show()
```

# Tarea
* Grafique los splines cúbicos tal que cumplan con la pendiente $m$ deseada en el punto $(x_1,y_1)$.

```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# Datos del problema
# =========================
x0, y0 = -1.0, 1.0
x1, y1 =  0.0, 5.0
x2, y2 =  1.0, 3.0
m = -2.0  # pendiente deseada en (x1,y1)

# =========================
# Spline por tramos:
# S0(x) = a0 + b0*(x-x0) + c0*(x-x0)^2 + d0*(x-x0)^3   en [x0,x1]
# S1(x) = a1 + b1*(x-x1) + c1*(x-x1)^2 + d1*(x-x1)^3   en [x1,x2]
# incógnitas: [a0 b0 c0 d0 a1 b1 c1 d1]
# =========================

def row_S0(x):
    t = x - x0
    return np.array([1, t, t**2, t**3, 0, 0, 0, 0], dtype=float)

def row_S1(x):
    u = x - x1
    return np.array([0, 0, 0, 0, 1, u, u**2, u**3], dtype=float)

def row_S0p(x):
    t = x - x0
    return np.array([0, 1, 2*t, 3*t**2, 0, 0, 0, 0], dtype=float)

def row_S1p(x):
    u = x - x1
    return np.array([0, 0, 0, 0, 0, 1, 2*u, 3*u**2], dtype=float)

def row_S0pp(x):
    t = x - x0
    return np.array([0, 0, 2, 6*t, 0, 0, 0, 0], dtype=float)

def row_S1pp(x):
    u = x - x1
    return np.array([0, 0, 0, 0, 0, 0, 2, 6*u], dtype=float)

# =========================
# Ecuaciones:
# 1) Interpolación:
#    S0(x0)=y0, S0(x1)=y1, S1(x1)=y1, S1(x2)=y2
# 2) Continuidad 2da derivada: S0''(x1)=S1''(x1)
# 3) Pendiente única: S0'(x1)=m y S1'(x1)=m
# 4) NATURAL DERECHA (para cerrar el sistema): S1''(x2)=0
# =========================

A = []
b = []

# Interpolación
A.append(row_S0(x0)); b.append(y0)
A.append(row_S0(x1)); b.append(y1)
A.append(row_S1(x1)); b.append(y1)
A.append(row_S1(x2)); b.append(y2)

# Continuidad de 2da derivada en x1: S0''(x1) - S1''(x1) = 0
A.append(row_S0pp(x1) - row_S1pp(x1)); b.append(0.0)

# Pendiente única en x1
A.append(row_S0p(x1)); b.append(m)
A.append(row_S1p(x1)); b.append(m)

# Natural derecha (cierra sistema)
A.append(row_S1pp(x2)); b.append(0.0)

A = np.vstack(A)
b = np.array(b, dtype=float)

coef = np.linalg.solve(A, b)
a0, b0, c0, d0, a1, b1, c1, d1 = coef

print("Coeficientes:")
print("S0: a0,b0,c0,d0 =", a0, b0, c0, d0)
print("S1: a1,b1,c1,d1 =", a1, b1, c1, d1)

def S0(x):
    t = x - x0
    return a0 + b0*t + c0*t**2 + d0*t**3

def S1(x):
    u = x - x1
    return a1 + b1*u + c1*u**2 + d1*u**3


# =========================
# Graficar
# =========================
xx0 = np.linspace(x0, x1, 200)
xx1 = np.linspace(x1, x2, 200)

plt.figure(figsize=(9, 6))
plt.plot(xx0, S0(xx0), label="S0(x) en [x0,x1]")
plt.plot(xx1, S1(xx1), label="S1(x) en [x1,x2]")

plt.scatter([x0, x1, x2], [y0, y1, y2], zorder=5)

# Tangente deseada
xt = np.linspace(x1 - 0.6, x1 + 0.6, 50)
yt = y1 + m*(xt - x1)
plt.plot(xt, yt, "--", label=f"Tangente deseada (m={m})")

plt.title("Spline cúbico con pendiente única en x1 (natural derecha)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import math

def spline_expr(i, a, b, c, d, xi, dec=6):
    """
    Devuelve un string: S_i(x)=a*(x-xi)^3 + b*(x-xi)^2 + c*(x-xi) + d
    Si algún coeficiente es NaN -> devuelve NaN en todos los casilleros.
    """
    coeffs = [a, b, c, d]
    if any((isinstance(v, float) and math.isnan(v)) for v in coeffs):
        return f"S{i}(x) = NaN*(x-{xi})^3 + NaN*(x-{xi})^2 + NaN*(x-{xi}) + NaN"

    # Redondeo para que no salga larguísimo
    a = round(float(a), dec)
    b = round(float(b), dec)
    c = round(float(c), dec)
    d = round(float(d), dec)
    xi = round(float(xi), dec)

    return f"S{i}(x) = {a}*(x-{xi})^3 + {b}*(x-{xi})^2 + {c}*(x-{xi}) + {d}"

print(spline_expr(0, a0, b0, c0, d0, x0))
print(spline_expr(1, a1, b1, c1, d1, x1))

```
