```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 3, 500)

y = x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

plt.figure(figsize=(8,5))
plt.plot(x, y, label='f(x)', color='blue')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True, linestyle='--', alpha=0.6)
plt.title('Gr√°fica del polinomio')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy import optimize

def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

intervalos = [ (-5, -2), (-2, 1), (1, 2)]
raices = []

for a, b in intervalos:
    try:
        r = optimize.bisect(f, a, b)
        if not any(np.isclose(r, x, atol=1e-6) for x in raices):
            raices.append(r)
    except ValueError:
        pass

raices = sorted(raices)

for i, r in enumerate(raices, 1):
    print(f"x{i} =", r)
```
